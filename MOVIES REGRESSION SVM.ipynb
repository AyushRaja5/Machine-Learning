# X-Y split

```python
x = df.drop('Collection',axis=1)
```

```python
y = df['Collection']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

## Standardise DATA

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train_stand = sc.fit_transform(X_train)
```

```python
X_test_stand = sc.fit_transform(X_test)
```

## Modeling 

```python
from sklearn.svm import SVR
svr_reg = SVR(kernel='linear' , C=1000)
```

```python
svr_reg.fit(X_train_stand,Y_train)
```

```python
Y_predict = svr_reg.predict(X_test_stand)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,r2_score
```

```python
r2_score(Y_test,Y_predict)
```
