```python
import cv2
img = cv2.imread('D:/imgs/seg.jpg')
from matplotlib import pyplot as plt
```

```python
print(img.shape)
cv2.imshow("original", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
imgheight=img.shape[0]
imgwidth=img.shape[1]
halfw = imgwidth//2
halfh = imgheight//2
```

```python
cropped_image1 = img[0:halfw, halfh:imgheight] # top right
cv2.imshow("top right", cropped_image1)
cv2.waitKey(0)
cv2.destroyAllWindows()

cropped_image2 = img[halfw:imgwidth, halfh:imgheight] # low Right
cv2.imshow("low Right", cropped_image2)
cv2.waitKey(0)
cv2.destroyAllWindows()

cropped_image3 = img[0:halfh, 0:halfw] # top left
cv2.imshow("top left", cropped_image3)
cv2.waitKey(0)
cv2.destroyAllWindows()

cropped_image4 = img[halfh:imgheight, 0:halfw] # Low Left
cv2.imshow("Low Left", cropped_image4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(7,7))
plt.subplot(1,1,1)
plt.imshow(img)
plt.title('Original Image')
```
