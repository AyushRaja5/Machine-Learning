```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow
```

```python
from skimage import io,feature
```

```python
img = io.imread("D:/imgs/footbal.jpg")
img_gray = io.imread("D:/imgs/footbal.jpg",as_gray=True)*255
```

```python
edge = feature.canny(img_gray,1,120)
```

```python
plt.subplot(1,2,1)
plt.imshow(img_gray,cmap='gray')
plt.title('Gray')
plt.subplot(1,2,2)
plt.imshow(edge,cmap='gray')
plt.title('Canny')
```
