```python
import cv2
import numpy as np
from skimage.util import random_noise
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread("D:/imgs/lpu.jpg")
# Add salt-and-pepper noise to the image.
noise_img = random_noise(img, mode='s&p',amount=0.3)
gnoise_img = np.array(255*noise_img, dtype = 'uint8')
```

```python
cv2.imshow('blur',gnoise_img)
cv2.waitKey(0)
```

```python
from skimage.restoration import denoise_tv_chambolle
from skimage import data, io

# Apply total variation filter denoising
denoised_image = denoise_tv_chambolle(noise_img,weight=0.1,multichannel=True)
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1,3,1)
plt.ylabel([4,2,0,6],fontsize=20,loc='center')
plt.imshow(noise_img)
plt.title('Original Image')

plt.subplot(1,3,2)
plt.imshow(denoised_image)
plt.title('Denoised image')
plt.xlabel([1,1,9,0],fontsize=20,loc='center')
```

```python
img_gray  = cv2.cvtColor(gnoise_img,cv2.COLOR_BGR2GRAY)
```
